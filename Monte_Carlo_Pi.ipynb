```python
import numpy as np
import matplotlib.pyplot as plt

x=np.random.uniform(-1, 1, 10000)
y=np.random.uniform(-1, 1, 10000)
points_inside=np.sum(x**2+y**2<=1)
points_total=len(x)
calc_pi=points_inside/points_total*4
print(f"Total points: {points_total}")
print(f"Points inside the circle: {points_inside}")
print(f"Calculated Pi: {calc_pi}")

plt.scatter(x, y, c=np.where(x**2+y**2<=1, 'red', 'blue'), s=1)
plt.axhline(y=0, color='gray')
plt.axvline(x=0, color='gray')
plt.axvline(x=-1, color='gray', linestyle='--')
plt.axvline(x=1, color='gray', linestyle='--')
plt.axhline(y=-1, color='gray', linestyle='--')
plt.axhline(y=1, color='gray', linestyle='--')
circle = plt.Circle((0, 0), 1, fill=False, color='red')
fig = plt.gcf()
ax = fig.gca()
ax.add_patch(circle)
plt.axis('equal')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.random.uniform(-1, 1, 10000)
y=np.random.uniform(-1, 1, 10000)
points_inside=np.sum(x**2+y**2<=1)
points_total=len(x)
calc_pi=points_inside/points_total*4
print(f"Total points: {points_total}")
print(f"Points inside the circle: {points_inside}")
print(f"Calculated Pi: {calc_pi}")

ycrop=float(input('Input ycrop where -1<ycrop<1: '))
cropped_area_points=np.sum((x**2+y**2<=1) & (y<=ycrop))
cropped_plot=cropped_area_points/points_inside*calc_pi
print(cropped_plot)

plt.scatter(x, y, c=np.where((x**2+y**2<=1) & (y<=ycrop), 'red', 'blue'), s=1)
plt.axhline(y=0, color='gray')
plt.axvline(x=0, color='gray')
plt.axvline(x=-1, color='gray', linestyle='--')
plt.axvline(x=1, color='gray', linestyle='--')
plt.axhline(y=-1, color='gray', linestyle='--')
plt.axhline(y=1, color='gray', linestyle='--')
plt.axhline(y=ycrop, color='black')
circle = plt.Circle((0, 0), 1, fill=False, color='red')
fig = plt.gcf()
ax = fig.gca()
ax.add_patch(circle)
plt.axis('equal')
plt.show()
```
