```python
import numpy as np
import matplotlib.pyplot as plt
import math
v0=200
dt=0.01
for angle in range(85,25,-5):
    theta = math.radians(angle)
    v0x=v0*math.cos(theta)
    v0y=v0*math.sin(theta)
    x_values=[]
    y_values=[]
    t=0.0
    while True:
        x=v0x*t
        y=v0y*t-0.5*9.8*t*t
        if y<0:
            break
        x_values.append(x)
        y_values.append(y)
        t+=dt     
    x_arr=np.array(x_values)
    y_arr=np.array(y_values)
    plt.plot(x_arr, y_arr)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math
v0=200
dt=0.01
damping=0.55
for angle in range(85,25,-10):
    theta = math.radians(angle)
    v0x=v0*math.cos(theta)
    v0y=v0*math.sin(theta)
    x_values=[]
    y_values=[]
    x=0
    y=0
    t=0.0
    tt=0.0
    while True:
        x=v0x*tt
        y=v0y*t-0.5*9.8*t*t
        if y<0:
            y=0
            v0y=v0y*damping
            t=0
            if v0y<50:
                break
        else:
            x_values.append(x)
            y_values.append(y)
            t+=dt  
            tt+=dt
    x_arr=np.array(x_values)
    y_arr=np.array(y_values)
    plt.plot(x_arr, y_arr)
plt.show()
```
