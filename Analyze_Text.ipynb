```python
import numpy as np
import re

file = open("text.txt")
content=file.read()
print(content)

unique_values, occurrence_count = np.unique(list(content), return_counts=True)
for value, count in zip(unique_values, occurrence_count):
    print(f"{value}: {count}")
print("")

words=np.array(content.split())
cleaned_words=np.array([re.sub(r'\W+', '', w.lower()) for w in words if w.strip() != ''])

unique_lengths, counts = np.unique(np.char.str_len(cleaned_words), return_counts=True)
for length, count in zip(unique_lengths, counts):
    print(f"{length} letter(s): {count} word(s)")
print("")

reversed_words=np.array([w[::-1] for w in cleaned_words])
palindromes=cleaned_words[cleaned_words==reversed_words]
palindromes=np.unique(palindromes)
lengths = np.char.str_len(palindromes)
sorted_by_length = palindromes[np.argsort(lengths)]
print("Palindromes in alphabetical order:")
print(np.sort(palindromes))
print("")
print("Palindromes in order of length:")
print(sorted_by_length)
print("")

unique_words, counts=np.unique(cleaned_words, return_counts=True)
sorted_by_count=unique_words[np.argsort(counts)[::-1]]
sorted_counts=np.sort(counts)
for u_word, count in zip(sorted_by_count, sorted_counts[::-1]):
    print(f"{u_word}: {count} time(s)")
print("")
```
